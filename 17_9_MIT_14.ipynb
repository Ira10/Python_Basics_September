#### Iteration vs Recursion

```python
## a*b using a while loop
def mult_iter(a, b):
    result = 0
    while b > 0:
        result += a
        b -= 1
    return result

print(mult_iter(5,4))
```

```python
## a*b recursive 
def mult_recur(a, b):
    if b == 1:
        return a
    else:
        return a + mult_recur(a, b-1)

print(mult_recur(5,4))
```

```python
############### YOU TRY IT #################
# Calculate n**p recursively by writing this function    MY SOLUTION
def power_recur(n, p):
    if p == 0 :
        return 1
    elif p == 1 :
        return n
    else:
        return n * power_recur(n, p-1)

print(power_recur(2,3))  # prints 8

print(power_recur(2,5))  # prints 8

print(power_recur(2,8))  # prints 8
```

#### Factorial iterative


```python
def factorial_iter(n):
    print(f'this is fact({n})')
    prod = 1
    for i in range(1,n+1):
        prod *= i
    return prod

factorial_iter(5)
```

### Factorial recursive


```python
def fact_recur(n):
    print(f'this is fact({n})')
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        # print(n*(n-1))
        return n * fact_recur(n-1)

fact_recur(5)
```
